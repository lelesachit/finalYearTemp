```python
import pandas as pd

# Function to convert vol strings to integers
def convert_vol_to_int(vol_str):
    if pd.isna(vol_str):
        return 1_000_000  # Default to 1M if NaN
    elif isinstance(vol_str, str) and 'M' in vol_str:
        return int(float(vol_str.replace('M', '')) * 1_000_000)
    elif isinstance(vol_str, str) and 'B' in vol_str:
        return int(float(vol_str.replace('B', '')) * 1_000_000_000)
    else:
        try:
            return int(vol_str)  # For values that are directly numbers
        except ValueError:
            return 1_000_000  # Default if something unexpected occurs
dfs = [df1, df2, df3, df4, df5, df6, df7, df8, df9]
# Loop through each DataFrame in dfs and apply the conversion
for df in dfs:
    for col in ['Vol_Nifty50', 'Vol_NiftyBank', 'Vol_NiftyFMCG', 'Vol_NiftyIT', 'Vol_NiftyMedia', 'Vol_NiftyMidcap', 'Vol_NiftyPharma', 'Vol_NiftyPSU', 'Vol_NiftySmallcap']:
        if col in df.columns:  # Check if the column exists in the DataFrame
            df[col] = df[col].apply(convert_vol_to_int)
```
